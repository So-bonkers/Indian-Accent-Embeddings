```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
from sklearn.cluster import KMeans
```

```python
# Load embeddings from file
file_name = "/home/sksystem/Downloads/phoneme_wav2vec_analysis/100 samples/aa_IITM.txt"
embeddings = np.loadtxt(file_name)
```

```python
# Apply k-means clustering
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters)
clusters = kmeans.fit_predict(embeddings)
```

```python
# Apply t-SNE to visualize clusters in 2D space
tsne = TSNE(n_components=3)
embeddings_2d = tsne.fit_transform(embeddings)
```

```python
# Plot clusters
plt.figure(figsize=(15, 15))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], cmap='viridis')
plt.title('t-SNE visualization of {} embeddings'.format(embeddings.shape[0]))
plt.colorbar()
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
# Load the data
data = np.loadtxt("/home/sksystem/Downloads/phoneme_wav2vec_analysis/100 samples/aa_libri.txt")

# Perform PCA
pca = PCA(n_components=3)
principal_components = pca.fit_transform(data)

# Perform K-means clustering
kmeans = KMeans(n_clusters=1, random_state=0).fit(principal_components)

# Visualize the clusters
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(principal_components[:,0], principal_components[:,1], principal_components[:,2], c=kmeans.labels_)
legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend1)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.show()
```

```python
# Load the data
data = np.loadtxt("/home/sksystem/Downloads/phoneme_wav2vec_analysis/100 samples/aa_libri.txt")

# Perform PCA
pca = PCA(n_components=3)
principal_components = pca.fit_transform(data)

# Perform K-means clustering
kmeans = KMeans(n_clusters=1, random_state=0).fit(principal_components)

# Visualize the clusters
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the first data with color from K-means clustering
scatter = ax.scatter(principal_components[:, 0], principal_components[:, 1], principal_components[:, 2], c=kmeans.labels_, label="Data 1")

# Load and plot the second data with a different color
data2 = np.loadtxt("/home/sksystem/Downloads/phoneme_wav2vec_analysis/100 samples/aa_IITM.txt")
principal_components2 = pca.transform(data2)
scatter2 = ax.scatter(principal_components2[:, 0], principal_components2[:, 1], principal_components2[:, 2], c='orange', label="Data 2")

# Set legends
legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend1)
legend2 = ax.legend([scatter, scatter2], ["Librispeech 3328 frames", "IITM 2271 frames"])

# Set labels
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")

# Show the plot
plt.show()
```
