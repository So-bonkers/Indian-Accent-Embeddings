```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# Load embeddings from file
file_name = "/home/sksystem/Downloads/train-clean-360/embeddings_librispeech/LibriSpeech/train-clean-360/combined_768.txt"
embeddings = np.loadtxt(file_name)
```

```python
# Apply k-means clustering
n_clusters = 45
kmeans = KMeans(n_clusters=n_clusters)
clusters = kmeans.fit_predict(embeddings)
```

```python
# Apply t-SNE to visualize clusters in 2D space
tsne = TSNE(n_components=3)
embeddings_2d = tsne.fit_transform(embeddings)
```

```python
# Plot clusters
plt.figure(figsize=(15, 15))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], c=clusters, cmap='viridis')
plt.title('t-SNE visualization of {} embeddings'.format(embeddings.shape[0]))
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Load the data
data = np.loadtxt("/home/sksystem/Downloads/train-clean-360/embeddings_librispeech/LibriSpeech/train-clean-360/combined_768.txt")
```

```python
# # Standardize the data
# data_std = (data - data.mean()) / data.std()
```

```python
# Perform PCA
pca = PCA(n_components=3)
principal_components = pca.fit_transform(data)
```

```python
# Perform K-means clustering
kmeans = KMeans(n_clusters=44, random_state=0).fit(principal_components)
```

```python
# Visualize the clusters
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(principal_components[:,0], principal_components[:,1], principal_components[:,2], c=kmeans.labels_)
legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend1)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.show()
```

```python
# Perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data)
```

```python
# Perform K-means clustering
kmeans = KMeans(n_clusters=44, random_state=0).fit(principal_components)
```

```python
# Visualize the clusters
fig, ax = plt.subplots()
scatter = ax.scatter(principal_components[:,0], principal_components[:,1], c=kmeans.labels_)
legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend1)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
